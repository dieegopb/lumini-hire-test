```python
# Bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 180)
import warnings
warnings.filterwarnings("ignore")
```

```python
# Carregando o dataset

df = pd.read_csv('Microdados_Enem_2016.csv')
df = df.drop(['Unnamed: 0', 'NU_INSCRICAO','CO_UF_RESIDENCIA', 'NU_ANO', 'CO_MUNICIPIO_RESIDENCIA','TP_ENSINO','CO_ESCOLA','NO_MUNICIPIO_ESC',
              'NO_MUNICIPIO_ESC','CO_UF_ESC','SG_UF_ESC','TP_DEPENDENCIA_ADM_ESC','TP_LOCALIZACAO_ESC','TP_SIT_FUNC_ESC','CO_MUNICIPIO_ESC',
              'NO_ENTIDADE_CERTIFICACAO','CO_UF_ENTIDADE_CERTIFICACAO', 'SG_UF_ENTIDADE_CERTIFICACAO'], axis =1)

df.head(10)
```

# 1. Informações sobre o dataset



```python
## 1.1  Informações referentes às variáveis categóricas.

df.describe(include=['O'])
```

```python
## 1.1  Informações referentes às variáveis numéricas.

df.describe()
```

# 2. Informações referente à prova objetiva

```python
# Filtrando o  dataset com o resultado das notas dos candidatos

df_provas = df[['NO_MUNICIPIO_RESIDENCIA','NO_MUNICIPIO_PROVA','SG_UF_RESIDENCIA', 'TP_COR_RACA','TP_ESCOLA','NU_IDADE','TP_SEXO','NU_NOTA_CN', 'NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT','TP_LINGUA','NU_NOTA_REDACAO']]
df_provas.dropna( inplace = True)

print('\n')
print( "Do total de candidatos inscristos no ENEM 2016, temos que  {}% compareceramm para a realização de todas as provas.".format( len(df_provas)/len(df)))
print('\n')
```

```python
# Gerando uma coluna a média das notas de cada candidato

df_provas['NU_NOTA_MEDIA'] = (df_provas['NU_NOTA_CN'] + df_provas['NU_NOTA_CH'] + df_provas['NU_NOTA_LC'] + df_provas['NU_NOTA_MT'] + df_provas['NU_NOTA_REDACAO'])/5
df_provas.head()
```

# 3 Gráficos

## 3.1 O gráfico a seguir  contém as médias e o desvio padrão para a media de notas em cada tipo de escola

```python
fig, ax = plt.subplots()

ind = np.arange(4) 
width = 0.35  
medias = list(df_provas.groupby('TP_ESCOLA')['NU_NOTA_MEDIA'].mean())
medias_std = list(df_provas.groupby('TP_ESCOLA')['NU_NOTA_MEDIA'].std())

plt.xlabel('Tipo de Escola')
plt.ylabel('Nota média por tipo de escola ')

p1 = ax.bar(ind, medias, width, bottom=0, yerr = medias_std)
# plt.xticks(ind , ('Não Respondeu', 'Pública', 'Privada', 'Exterior' ))
# plt.show()
```


## 3.2 O gráfico a seguir  contém as médias e o desvio padrão para a media de notas em cada tipo de de cor/raça

```python
fig, ax = plt.subplots()

ind = np.arange(6) 
width = 0.35  
medias = list(df_provas.groupby('TP_COR_RACA')['NU_NOTA_MEDIA'].mean())
medias_std = list(df_provas.groupby('TP_COR_RACA')['NU_NOTA_MEDIA'].std())

plt.xlabel('Tipo de Escola')
plt.ylabel('Nota média por tipo de escola ')

p1 = ax.bar(ind, medias, width, bottom=0, yerr = medias_std)
plt.xticks(ind , ('Não declarado','Branca','Preta','Parda','Amarela','Indígena'))
# plt.show()
```

## 3.3 O pŕoximo gráfico mostrará como a idade dos participantes está distribuída

```python
plt.figure(figsize=(12,4))
plt.hist(df['NU_IDADE'][df['NU_IDADE'].notna()], bins='auto', normed= True)
plt.grid(True)
plt.xlabel('Idade fos candidatos')
plt.ylabel('Porcentagem dos candidados por idade')
plt.show()
```

## 3.3 Boxplot com as médias das notas do ENEM por estado.

```python
### Boxplot com as médias das notas por estado.

plt.figure(figsize=(12,4))
ax = sns.boxplot(x="SG_UF_RESIDENCIA", y="NU_NOTA_MEDIA", data=df_provas)

plt.xlabel('Estados')
plt.ylabel('NU_NOTA_MEDIA')
```

## 3.4 Boxplot com as médias das notas da redação do ENEM por estado.


```python
### Boxplot com a médias da  nota de redação por estado.

plt.figure(figsize=(12,4))
ax = sns.boxplot(x="SG_UF_RESIDENCIA", y="NU_NOTA_REDACAO", data=df_provas)

plt.xlabel('Estados')
plt.ylabel('NU_NOTA_REDACAO')
```

# 4. Identificando os municípios que obtiveram, em média, as menores e maiores médias na prova do ENEM 2016


### 4.1 Municípios  que obtiveram, em média, médias na prova abaixo que 400

```python
df_Nota_Media_municípios = pd.DataFrame(df_provas.groupby('NO_MUNICIPIO_PROVA')['NU_NOTA_MEDIA'].mean())
df_menor_400 = df_Nota_Media_municípios[df_Nota_Media_municípios['NU_NOTA_MEDIA']<400]
print(df_menor_400)
```

### 4.2 Municípios  que obtiveram, em média, médias na prova maior que  650

```python
df_maior_650 = df_Nota_Media_municípios[df_Nota_Media_municípios['NU_NOTA_MEDIA']>650]
print(df_maior_650)
```
